# Задача 1

Разделить таблицу данных так, чтобы в обучающую выборку попало 66% наблюдений, а в тестовую выборку 33% наблюдений.

```python
import pandas as pd
import numpy as np
```

```python
# исходная таблица
df = pd.read_csv("./data/beverage_r.csv", sep = ';').drop(columns = 'numb.obs')
df
```

```python
# Число строк в таблице
len(df)
```

```python
# перемешиваем индексы
np.random.seed(seed = 42)
indices = np.random.permutation(len(df))
indices
```

```python
train_size = int(0.66 * len(df))

# тренировочная выборка
train_df = df.iloc[indices[:train_size]]
train_df
```

```python
# тестовая выборка
test_df = df.iloc[indices[train_size:]]
```

```python
# соотношение размеров
f'{len(train_df) / len(df):.2f}, {len(test_df) / len(df):.2f}'
```

# Задача 2

Разделить таблицу данных на четыре одинаковых части.

```python
r1 = np.random.uniform(low = 0, high = 1.0, size = len(df))

# 3 квантиля
q1 = np.quantile(r1, 0.25)
q2 = np.quantile(r1, 0.50)
q3 = np.quantile(r1, 0.75)
q1, q2, q3
```

```python
# разделям данные по квартилям
part1 = df.iloc[r1 < q1]
part2 = df.iloc[(r1 >= q1) & (r1 < q2)]
part3 = df.iloc[(r1 >= q2) & (r1 < q3)]
part4 = df.iloc[r1 >= q3]
```

# Задача 3

Разделить таблицу данных на две стратифицированные подвыборки.
По переменным кока-кола (**COKE**) и минералка (**TAB**).

```python
df['strata'] = df['COKE'].astype(str) + '_' + df['TAB'].astype(str)
df
```

```python
# индексы тренировочных и тестовых строк
train_indices, test_indices = [], []
```

```python
strata_unique = df['strata'].unique()

for stratum in strata_unique:
    stratum_indices = df[df['strata'] == stratum].index.to_list()
    
    np.random.shuffle(stratum_indices)

    mid = len(stratum_indices) // 2

    train_indices.append(stratum_indices[:mid])
    test_indices.append(stratum_indices[mid:])
```

```python
from itertools import chain

train_indices = list(chain(*train_indices))
test_indices = list(chain(*test_indices))
```

```python
# тренировочная выборка
train_df = df.loc[train_indices].drop('strata', axis = 1)
train_df
```

```python
# тестовая выборка
test_df = df.loc[test_indices].drop('strata', axis = 1)
train_df
```
