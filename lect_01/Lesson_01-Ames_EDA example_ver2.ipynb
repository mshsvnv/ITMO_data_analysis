### Ядерные оценки плотности

```python
#  Активируем библиотеку, 
#  в которой есть команды для построения ядерных оценок плотности

from scipy.stats.kde import gaussian_kde

from numpy import linspace,hstack
from pylab import plot,show,hist

#  отсутствует sheather jones bandwidth
#  присутствуют только Scott Silverman normal density reference
```

```python
#  вручную подбираем ширину окна

my_density = gaussian_kde(AH['SalePrice'])

#  my_density = gaussian_kde(AH['SalePrice'], bw_method = 5)
#  my_density = gaussian_kde(AH['SalePrice'], bw_method = 1)
#  my_density = gaussian_kde(AH['SalePrice'], bw_method =0.05)

# my_density = gaussian_kde(AH['SalePrice'], bw_method =0.1)


# график ядерной оценки плотности
x = linspace(min(AH['SalePrice']), max(AH['SalePrice']),1000)
plot(x, my_density(x),'g')                    # оценка плотности

hist(AH['SalePrice'], density=1, alpha=.3)    # добавляем гистограмму
show()
```

```python
#  график ядерной оценки плотности

plot(x, my_density(x),'r') 
```

###  Добавление о  диаграммах рассеивания

```python
#   https://python-graph-gallery.com
plt.style.use('seaborn')
```

```python
 
# Генерируем данные - три кластера
# Три нормально распределенных облака, координаты независимы
df=pd.DataFrame({'x_1': np.random.normal(10, 1.2, 20000), 'x_2': np.random.normal(10, 1.2, 20000), 'group': np.repeat('A',20000) })
tmp1=pd.DataFrame({'x_1': np.random.normal(14.5, 1.2, 20000), 'x_2': np.random.normal(14.5, 1.2, 20000), 'group': np.repeat('B',20000) })
tmp2=pd.DataFrame({'x_1': np.random.normal(9.5, 1.5, 20000), 'x_2': np.random.normal(15.5, 1.5, 20000), 'group': np.repeat('C',20000) })
# Склеиваем три таблицы в одну
df=df.append(tmp1).append(tmp2)
 
# диаграмма рассеивания
plt.plot( 'x_1', 'x_2', data=df, linestyle='', marker='o')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Слишком крупные точки:', loc='left')
```

```python
# диаграмма рассеивания
plt.plot( 'x_1', 'x_2', data=df, linestyle='', marker='o', markersize=0.7)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Маленькие точки, но не всегда так получается', loc='left')
```

```python
# Ядерная оценка плотности
#  Обратите внимание, как долго считается


with sns.axes_style('white'):
    sns.jointplot('x_1', 'x_2', data=df, kind='kde')
    
   
    
```

```python
#  2D гистограмма, основания - 6-угольники
#  Вычисления намного быстрее

with sns.axes_style('white'):
    sns.jointplot('x_1', 'x_2', data=df, kind='hex')
```
